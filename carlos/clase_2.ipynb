```python
students = [
    {
        "name": "valentina",
        "lastname": "arenas",
        ("classes", 0): ["python", "scala"]
    },
    {
        "name": "juan",
        "lastname": "perez",
        ("classes", 0): ["python", "java"],
        True: "algo"
    }
]
students
```

```python
print(type(students))
print(students[0])
print(type(students[0]))
print(students[0].keys())
print(type(students[0].keys()))
print(list(students[0].keys())[2])
print(type(list(students[0].keys())[2]))
print(students[0][('classes', 0)])
print(type(students[0][('classes', 0)]))
print(students[0][('classes', 0)][0])
```

```python
# listas/diccionarios comprensivos
listica = []

for j in range(5):
    for i in range(5):
        if i*j % 2 == 0:
            listica.append(i*j)

print(listica)
print(len(listica))

listica = [i*j for j in range(5) for i in range(5) if i*j % 2 == 0]
print(listica)
print(len(listica))
```
