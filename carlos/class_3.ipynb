```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.classes = []

    def add_class(self, class_name):
        self.classes.append(class_name)
        print(f"{self.name} se inscribió en la clase de: {class_name}")
```

```python
# s = Student(25, "Valentina")
s = Student(age=25, name="Valentina")
s.add_class("Python")
s.add_class("Pandas")
s.add_class("Matplotlib")

c = Student(age=25, name="Carlos")
c.add_class("Python")
c.add_class("Pandas")
c.add_class("Matplotlib")
c.add_class("Git")
c.classes
```

```python
print(s.classes)
print(c.classes)
```

# pandas!

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Step 1: Create a dataset with random data
# Create a DataFrame with pandas
np.random.seed(0)  # For reproducibility
data = {
    'A': np.random.randint(1, 100, 10),
    'B': np.random.randint(1, 100, 10),
    'C': np.random.randint(1, 100, 10)
}
df = pd.DataFrame(data)
df
```

```python
lista = [1,2,3]
print(lista + lista)
print(min(lista))

lista = np.array([1,2,3])
print(lista + lista)
print(min(lista))
```

```python
# Step 2: Calculate some statistics with numpy
mean_A = np.mean(df['A'])
mean_B = np.mean(df['B'])
mean_C = np.mean(df['C'])

print(f"\nMean of column A: {mean_A}")
print(f"Mean of column B: {mean_B}")
print(f"Mean of column C: {mean_C}")

print(df["A"].mean())
print(df["A"].min())
print(df["A"].max())
print(df["A"].unique())
print(df["A"].nunique())
```

```python
# Step 3: Plot the data with matplotlib
plt.figure(figsize=(8, 5))

# Plotting each column
plt.plot(df.index, df['A'], label='Column A', marker='o')
plt.plot(df.index, df['B'], label='Column B', marker='s')
plt.plot(df.index, df['C'], label='Column C', marker='^')

# Adding titles and labels
plt.title('Random Data Visualization')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend()

# Display the plot
plt.show()
```

```python
df.rename(columns={
    "Codigo_Fuente_Comprobante": "no_factura",
    "Periodo_aomes": "periodo",
    "Nombre_Bodega": "tienda",
    "Suma de Cantidad_Unidades": "unidades",
    "Suma de Valor_Neto": "valor_neto"
    },
    inplace=True
)
df.columns
```

```python
plt.figure(figsize=(10, 6))

for t in df["tienda"].unique():
    _filter = df["tienda"] == t
    tmp = df[_filter]
    group = tmp.groupby(by="periodo").agg({'unidades': 'sum', 'valor_neto': 'sum'})
    plt.plot(group.index, group['valor_neto'], marker='o', label=t)

plt.show()
```
